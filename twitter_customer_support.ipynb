```python
def preprocess_text(text):
    # Converting text to string and Lowercasing
    text = str(text).lower()

    # Tokenization (used TweeteTokenizer to preprocess tweets effectively)
    tokens = tokenizer.tokenize(text)

    # Removing noise
    tokens = [re.sub(r'http\S+', '', token) for token in tokens]  # Remove URLs

    tokens = [re.sub(r'@[A-Za-z0-9_]+', '', token) for token in tokens]  # Remove mentions

    tokens = [re.sub(r'#', '', token) for token in tokens]  # Remove hashtags

    tokens = [re.sub(r'[^\w\s]', '', token) for token in tokens]  # Remove punctuation and special characters

    tokens = [token for token in tokens if token not in stop_words]  # Remove stopwords

    tokens = [abbr_dict[token] if token in abbr_dict else token for token in tokens]    # Remove slang terms

    tokens = [lemmatizer.lemmatize(token) for token in tokens]  # Lemmatization

    tokens = [stemmer.stem(token) for token in tokens] # Stemming

    return ' '.join(tokens)

```

```python
# Apply preprocessing to all rows
df1['text'] = df1['text'].apply(preprocess_text)
```
