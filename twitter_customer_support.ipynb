```python
from textblob import TextBlob

"""
preprocess text function to remove URLs,mentions
punctuation and special characters, hashtags ,stopwords and
do stemming,lemmetization and spelling correction
"""

def preprocess_text(text):
    # Converting text to string and Lowercasing
    text = str(text).lower()

    # Tokenization (used TweeteTokenizer to preprocess tweets effectively)
    tokens = tokenizer.tokenize(text)

    # Removing noise
    tokens = [token for token in tokens if token.isalpha()]  # Keep only alphabetic tokens

    tokens = [re.sub(r'http\S+', '', token) for token in tokens]  # Remove URLs

    tokens = [re.sub(r'@[A-Za-z0-9_]+', '', token) for token in tokens]  # Remove mentions

    tokens = [re.sub(r'#', '', token) for token in tokens]  # Remove hashtags

    tokens = [re.sub(r'[^\w\s]', '', token) for token in tokens]  # Remove punctuation and special characters


    tokens = [token for token in tokens if token not in stop_words]  # Remove stopwords

    tokens = [abbr_dict[token] if token in abbr_dict else token for token in tokens]    # Remove slang terms

    tokens = [lemmatizer.lemmatize(token) for token in tokens]  # Lemmatization

    tokens = [stemmer.stem(token) for token in tokens] # Stemming

    #Correct spelling
    tokens = [str(TextBlob(token).correct()) for token in tokens]

    return ' '.join(tokens)

```

# **One-Hot Encoding (OHE):**

**Advantages:**

* Simple and straightforward representation of categorical data.

* Retains all information about the presence or absence of each category.

* Compatible with various machine learning algorithms.

* Suitable for cases where the categorical variables have a small number of unique values.

* Works well when the order or frequency of categories is not important.





**Disadvantages:**

* Generates high-dimensional sparse vectors, which can be memory-intensive for large datasets.
* Doesn't capture the semantic meaning or similarity between categories.
* Doesn't handle unseen categories well without additional preprocessing.
* Treats all categories equally, regardless of their importance or relevance.
* May result in overfitting when dealing with high-dimensional data.

# **Bag-of-Words (BOW):**

**Advantages:**

* Simple and efficient representation of text data.
* Captures the frequency of words in the document.
* Preserves the order of words in the document to some extent.
* Allows for the extraction of important features from text data.
* Works well for tasks like sentiment analysis, text classification, and information retrieval.

**Disadvantages:**

* Ignores the semantic meaning and context of words.
* Doesn't handle misspellings, synonyms, or variations of words well.
* Can lead to high-dimensional sparse matrices, especially with a large vocabulary.
* Doesn't consider the grammatical structure or syntax of sentences.
* Treats all words equally, regardless of their importance or relevance to the document.

# **N-grams:**

**Advantages:**

* Captures contextual information by considering sequences of words rather than individual words.
* Provides better context and understanding of language compared to BOW.
* Allows for the extraction of phrases and idiomatic expressions from text data.
* Helps in capturing local word dependencies and patterns.
* Can improve the performance of models in tasks like language modeling, machine translation, and speech recognition.

**Disadvantages:**

* Increases the dimensionality of the feature space, leading to sparsity and computational complexity.
* Requires more memory and computational resources compared to BOW, especially for higher-order n-grams.
* May suffer from data sparsity issues, especially for higher-order n-grams.
* More susceptible to noise and irrelevant information, especially for longer n-grams.
* Doesn't capture long-range dependencies or relationships between distant words effectively.

# **Term Frequency-Inverse Document Frequency (TF-IDF):**
**Advantages:**

* Highlights the importance of words based on their frequency in the document and rarity across the corpus.
* Reduces the importance of common words that occur in many documents (e.g., stopwords).
* Provides a more nuanced representation of text data compared to BOW.
* Helps in identifying key terms or features that are discriminative for each document.
* Effective in handling large vocabularies and sparse matrices.

**Disadvantages:**

* Doesn't capture the semantic meaning or context of words, similar to BOW.
* Requires tuning of parameters such as the smoothing factor and document frequency threshold.
* May not perform well for short documents or in cases where the corpus is small.
* Loss of information about word order and sentence structure, similar to BOW.
* Doesn't handle misspellings, synonyms, or variations of words well, similar to BOW.
